```python
import pandas as pd
import os
from sklearn import metrics
import numpy as np
from sklearn.metrics import accuracy_score 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
heartFailure = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
```

```python
heartFailure.head()
```

```python
heartFailure.tail()
```

```python
features = ['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']
labels = 'DEATH_EVENT'
```

```python
train, test = train_test_split(heartFailure, test_size = 0.25, random_state = 1)
trainX = train[features]
trainY = train[labels]
testX = test[features]
testY = test[labels]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=34, max_depth=9)
rf.fit(trainX, trainY)
y_pred = rf.predict(testX)
print("Accuracy: ", metrics.accuracy_score(testY, y_pred))
```

```python
predict = {}

print("Please type the integers as prompted")
for feature in features:
  val = input(feature + ":")
  predict[feature] = val 

predict = pd.DataFrame(predict, index=[0])

print 

output = rf.predict(predict)
```

```python
print("Model Prediction of \'DEATH EVENT\' — ", end = "")
print(output)
print()

if output == [1]:
  print("Based on the model, this data indicates a HIGH-RISK of heart failure.")
if output == [0]:
  print("Based on the model, this data indicates a LOW-RISK of heart failure.")
```
